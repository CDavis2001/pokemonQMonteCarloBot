```python
file = open("qvaluetracking/QLite.txt", "r")
data = dict()
states = []

line_num = 0
while True:
    line_num += 1
    if line_num % 10000 == 0:
        print(line_num)
        
    line = file.readline()
    if not line:
        break
    point = line.split(";")
    if len(point) != 3:
        continue
    if not point[0] in states:
        states.append(point[0])
        data[point[0]] = [float(point[1]), float(point[2].strip())]
    else:
        if point[1] == point[2]:
            continue
        else:
            data[point[0]].append(float(point[2].strip()))
                
    
file.close()
longest = [[]]*100
lengths= [0]*100
count = 0
for key in data:
    longest.append(data[key])

file = open("longest.txt", "w")
for i in range(10):
    maxList = max(longest, key= len)
    file.write(str(maxList))
    file.write("\n")
    longest.remove(maxList)

file.close()        


    
```
